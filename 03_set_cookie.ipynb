```python
import random
```

```python
class Colors:
    RED = '\033[31m'
    BLUE = '\033[34m'
    GREEN = '\033[32m'
    RESET = '\033[0m'
```

```python
import socket

host='localhost'
port=8000


def generate_response(cookie_val):
    response = f"""HTTP/1.1 200 OK\r
Content-Type: text/html\r
Set-Cookie: a_cookie={cookie_val}\r
\r
<h1>Hello, World!</h1>\r
    """
    return response


# Create a socket object
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Bind the socket to a specific address and port
server_address = (host, port)
server_socket.bind(server_address)

# Listen for incoming connections (only allow one connection at a time for simplicity)
server_socket.listen(1)  #backlog of 1

print(f"Server listening on {host}:{port}")
print(f"Open a browser, type {host}:{port} in the URL field, and press enter.\n")

try:
    while True:
        # Accept a connection
        connection, client_address = server_socket.accept()
        print(f"{Colors.BLUE}Connection from: {client_address}{Colors.RESET}")
    
        try:
            # Receive the data
            data = connection.recv(2048) # receive up to 1024 bytes
            print(data)

            cookie_val = random.randint(0, 100)
            print(f"Setting cookie to: {cookie_val}")
            response = generate_response(cookie_val)
            connection.sendall(response.encode('utf-8'))
            print(f"\n{Colors.GREEN}Sent response:{Colors.RESET}")
            print(response)
        finally:
            # Clean up the connection
            connection.close()
except KeyboardInterrupt:  #Handle ctrl+c to close the server cleanly.
    print("\nServer shutting down...")
finally:
    server_socket.close() #close the socket
    print("Server stopped.")
```
