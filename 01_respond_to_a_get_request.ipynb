```python
import socket

host='localhost'
port=8000
"""
Simple HTTP server that listens for a single request, prints the request,
and sends a simple "Hello, World!" response.
"""

# Create a socket object
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Bind the socket to a specific address and port
server_address = (host, port)
server_socket.bind(server_address)

# Listen for incoming connections (only allow one connection at a time for simplicity)
server_socket.listen(1)  #backlog of 1

print(f"Server listening on {host}:{port}")
print(f"Open a browser, type {host}:{port} in the URL field, and press enter.\n")

try:
    # Accept a connection
    connection, client_address = server_socket.accept()
    print(f"Connection from: {client_address}")

    try:
        # Receive the data
        data = connection.recv(1024) # receive up to 1024 bytes
        if data:
            decoded_data = data.decode('utf-8') # Decode from bytes to string
            print("Received data:")
            print(decoded_data)

            # Construct a simple HTTP response
            response = "HTTP/1.1 200 OK\r\n"  #Status line.  Important to end lines with \r\n
            response += "Content-Type: text/html\r\n" #Content-Type Header.   Important to end lines with \r\n
            response += "\r\n"                       #Blank Line: Separates headers from body. Important to end lines with \r\n
            response += "<html><body><h1>Hello, World!</h1></body></html>\r\n" #The HTML Body

            # Send the response back to the client (encode back to bytes)
            connection.sendall(response.encode('utf-8'))
            print("\nSent response:")
            print(response)

    finally:
        # Clean up the connection
        connection.close()

except KeyboardInterrupt:  #Handle ctrl+c to close the server cleanly.
    print("\nServer shutting down...")

finally:
    server_socket.close() #close the socket
    print("Server stopped.")
```
